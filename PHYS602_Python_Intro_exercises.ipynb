# 3  Logic and Conditions

Logic operators compare two values and return `True` or `False` (a **boolean**). They form the building blocks of decision-making in code.

| Operator | Meaning | Example | Result |
|----------|---------|---------|--------|
| `==` | Equal | `1 == 1` | `True` |
| `!=` | Not equal | `1 != 2` | `True` |
| `<` | Less than | `1 < 2` | `True` |
| `<=` | Less than or equal | `2 <= 2` | `True` |
| `>` | Greater than | `3 > 2` | `True` |
| `>=` | Greater than or equal | `1 >= 2` | `False` |

**Combining conditions:**
* `and` — both must be `True`
* `or` — at least one must be `True`
* `not` — negates (flips) the boolean

>  Do NOT confuse `=` (assignment) with `==` (comparison)!


````python
# ── Exercise 3.1 ──────────────────────────────────────────────────
a = 1
b = 2

print(f"a == b : {a==b    }")
print(f"a != b : {  a!=b  }")
print(f"a < b  : {  a<b  }")
print(f"a <= b : {  a<=b  }")
print(f"a > b  : {  a>b  }")
print(f"a >= b : {  a>=b  }")

# Combining with and / or
c = 3
print(f"a < b and b < c  : {  a < b and b < c  }")   # True and True
print(f"a < b and b > c  : { a < b and b > c   }")   # True and False
print(f"a > b or  b < c  : {  a > b or  b < c   }")   # False or True
print(f"a > b or  b > c  : {  a > b or  b > c  }")   # False or False

# Negation
print(f"not (a < b)      : {  not (a < b)       }")
````

## 4.2  while Loops

A `while` loop repeats a block **as long as its condition remains `True`**:

```python
counter = 0
while counter < 5:
    print(counter)
    counter += 1   # short for: counter = counter + 1
```

> If the condition never becomes `False` you get an **infinite loop**. Always make sure something inside the loop moves you towards the exit condition.

**Loop control keywords:**
* `break` — exit the loop immediately.
* `continue` — skip the rest of this iteration and go to the next.


````python
# ── Exercise 4.2 ──────────────────────────────────────────────────
# (a) Complete the while loop to count from 0 to 9
a = 0
while  a<9 :      
    print(f"a is {a}")
    a = a + 1

# (b) Use continue and break:
#     Skip printing when a==5, stop entirely when a==7.
a = 0
while a < 10:
    a += 1
    if a == 5:
        continue
        pass
    if a == 7:
        break
        pass
    print(f"a is {a}")

# (c) NEW — Compute the factorial of n using a while loop (n! = 1*2*…*n)
n = 6
result =1
index=-1
while index<n-1:
    index=index+1
    result= result*(n-index)

print(f"{n}! = {result}")   # expected: 720
````

When we need to store many related values together, Python offers several **collection** types. The two most important for scientific work are **lists** and **dictionaries**.


| `my_list.insert(i, x)` | Insert `x` at index `i` |
| `len(my_list)` | Return the number of elements |

# 6  Functions

A **function** is a named, reusable block of code. It can accept **arguments** (input values) and optionally **return** a result. Functions are the primary way we avoid writing the same code twice.

# 7  Error Handling  *(New)*

In Python, runtime errors are called **exceptions**. If an exception is not handled, the program crashes with a traceback. We can gracefully catch exceptions using `try / except`:

```python
try:
    result = 10 / 0          # raises ZeroDivisionError
except ZeroDivisionError:
    print("Cannot divide by zero!")
```

**Common exception types:**

| Exception | Cause |
|-----------|-------|
| `ZeroDivisionError` | Division by zero |
| `ValueError` | Wrong type or value (e.g. `int("hello")`) |
| `TypeError` | Operation on incompatible types |
| `IndexError` | List index out of range |
| `KeyError` | Dictionary key not found |

You can catch multiple exceptions, run cleanup code with `finally`, and raise your own exceptions with `raise`:

```python
try:
    x = int(input("Enter a number: "))
except ValueError as e:
    print(f"That wasn't a number: {e}")
except ZeroDivisionError:
    print("Cannot divide by zero")
else:
    print(f"Success! x = {x}")    # runs only if no exception was raised
finally:
    print("This always runs.")     # cleanup code
```


````python
# ── Exercise 7.1 — try / except ─────────────────────────────────
# (a) Wrap the float conversion in a try/except to handle ValueError
values = ["3.14", "hello", "2.71", "NaN_string", "0"]
results = []
for v in values:
    try:
        results.append( float(v)   )   # YOUR CODE HERE — convert v to float
    except   ValueError :                # YOUR CODE HERE — catch the right exception
        results.append(None)
print(results)   # expected: [3.14, None, 2.71, None, 0.0]

# (b) Safe divide function — return None and print a warning if dividing by zero
def safe_divide(a, b):
    try:
        c= a/b
        return c
    except:
        print('Divided by zero')
        return None

print(safe_divide(10, 2))   # 5.0
print(safe_divide(10, 0))   # None + warning

# (c) NEW — Write a function safe_index(lst, i) that returns lst[i] or
#     None (with a message) if the index is out of range.
def safe_index(lst, i):
    try:
        return lst[i]
    except IndexError:
        print("Warning: index out of range")
        return None

my_list = [10, 20, 30]
print(safe_index(my_list, 1))   # 20
print(safe_index(my_list, 9))   # None + warning
````

## 8.1  Plotting with Matplotlib

The basic Matplotlib workflow:

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.figure(figsize=(6, 4))   # optional: set figure size in inches
plt.plot(x, y, label="x²")
plt.xlabel("x")
plt.ylabel("y")
plt.title("My first plot")
plt.legend()
plt.grid()
plt.show()
```

Common plot types: `plt.plot()` (line), `plt.scatter()` (scatter), `plt.bar()` (bar chart), `plt.hist()` (histogram).

## 8.2  NumPy Arrays

A NumPy **array** looks like a list but:
1. All elements must be the **same type**.
2. Operations apply **element-wise** — no explicit `for` loop needed.
3. Maths is done in optimised C code, making it orders of magnitude faster.

```python
a = np.array([1, 2, 3, 4])
a + 10          # → [11, 12, 13, 14]
a * 2           # → [2, 4, 6, 8]
np.sqrt(a)      # → [1, 1.414, 1.732, 2]
```

Creation shortcuts:
```python
np.zeros(5)              # [0. 0. 0. 0. 0.]
np.ones((3, 3))          # 3×3 matrix of 1s
np.linspace(0, 1, 5)     # 5 evenly spaced points between 0 and 1
np.arange(0, 10, 2)      # [0, 2, 4, 6, 8]  (like range but returns array)
```


````python
# ── Exercise 8.3 — NumPy Arrays ─────────────────────────────────
# (a) Array arithmetic — try adding 1 to a plain list (will fail!)
my_list = [1, 2, 3, 4]
try:
    bad = my_list + 1
except TypeError as e:
    print(f"TypeError: {e}")

# (b) Convert the list to a numpy array and repeat
my_array = np.array(my_list)
print(my_array + 1)    # works!
print(my_array ** 2)

# (c) Array operations
a = np.linspace(-5, 5, 50)
b =  a**2  # YOUR CODE HERE — a^2
c =  -a**2+10  # YOUR CODE HERE — -a^2 + 10

plt.plot(a, b, label="a²")
plt.plot(a, c, label="-a² + 10")
plt.xlabel("a")
plt.legend()
plt.grid()
plt.title("NumPy array operations")
plt.show()

# (d) Butterfly curve — parametric plot
t = np.linspace(0, 12 * np.pi, 2000)
x =  np.sin(t)*(np.exp(np.cos(t))-2*np.cos(4*t)+np.sin(t/12)**5)  # YOUR CODE HERE: sin(t) * [exp(cos(t)) - 2cos(4t) + sin^5(t/12)]
y =   np.cos(t)*(np.exp(np.cos(t))-2*np.cos(4*t)+np.sin(t/12)**5) # YOUR CODE HERE: cos(t) * [exp(cos(t)) - 2cos(4t) + sin^5(t/12)]
plt.plot(x, y)
plt.title("Butterfly Curve")
plt.axis("equal")
plt.show()
````

````python
# ── Exercise 8.4 — Plotting Exercises ───────────────────────────
# (a) Plot the cubic polynomial y = x³ + 4.5x² + 7x - 1  on  x ∈ [-10, 10]
a_coef, b_coef, c_coef, d_coef = 1, 4.5, 7, -1
x = np.linspace(-10,10, 200)   # YOUR CODE HERE — np.linspace
y =  a_coef*x**3+b_coef*x**2+c_coef*x+d_coef

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Cubic polynomial")
plt.grid()
plt.show()

# (b) Power-law with exponential cut-off: f(x) = n * x^(-g) * exp(-x/c)
#     n=1e-12, g=2.5, c=10,  x ∈ [0.001, 100] — use log scale axes
n_norm, g, c = 1e-12, 2.5, 10
x = np.logspace(np.log10(0.001),np.log10(100), 300)
y = n_norm*x**(-g)*np.exp(-x/c)

plt.plot(x, y)
plt.xscale("log")
plt.yscale("log")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Power law with exponential cut-off")
plt.grid(which="both", ls="--", alpha=0.5)
plt.show()

# (c) NEW — Find the x-intersection of y1 = 0.2x + 7 and y2 = 1.5x - 3
#     numerically using np.argmin on the absolute difference
x = np.linspace(-10, 10, 10000)
y1 = 0.2 * x + 7
y2 = 1.5 * x - 3

idx =  np.argmin(np.abs(y1-y2))
x_int = x[idx]
print(f"Intersection at x ≈ {x_int:.4f}")   # expected ≈ 7.692

plt.plot(x, y1, label="y1")
plt.plot(x, y2, label="y2")
plt.axvline(x_int, color="k", ls="--", label=f"x ≈ {x_int:.2f}")
plt.legend()
plt.grid()
plt.show()
````
